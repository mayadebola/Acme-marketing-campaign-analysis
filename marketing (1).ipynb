```python
# Strip leading and trailing spaces from column names
marketing_data.columns = marketing_data.columns.str.strip()

# Now remove the dollar sign and commas from the 'Income' column
marketing_data['Income'] = marketing_data['Income'].str.replace('$', '', regex=True)
marketing_data['Income'] = marketing_data['Income'].str.replace(',', '', regex=True)

# Convert the column to numeric, coercing errors into NaN
marketing_data['Income'] = pd.to_numeric(marketing_data['Income'], errors='coerce')

# Check the data types and the first few rows
print(marketing_data.dtypes)
print(marketing_data.head())
```

```python
##Customer Segmentation
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#### Purchasing Behavior Analysis
#Analyzing spending patterns across different product categories
# List of spending columns
spending_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

#the total spending per customer
marketing_data['TotalSpending'] = marketing_data[spending_columns].sum(axis=1)

#spending patterns across categories
spending_means = marketing_data[spending_columns].mean()
spending_totals = marketing_data[spending_columns].sum()

# Plot of spending patterns
plt.figure(figsize=(14, 7))

# Plot for average spending per category
plt.subplot(1, 2, 1)
sns.barplot(x=spending_means.index, y=spending_means.values, palette='viridis')
plt.xlabel('Product Category')
plt.ylabel('Average Spending')
plt.title('Average Spending per Product Category')
plt.xticks(rotation=45)

# Plot for total spending per category
plt.subplot(1, 2, 2)
sns.barplot(x=spending_totals.index, y=spending_totals.values, palette='viridis')
plt.xlabel('Product Category')
plt.ylabel('Total Spending')
plt.title('Total Spending per Product Category')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# top-performing products
top_products = spending_totals.sort_values(ascending=False)
print("Top-Performing Products:")
print(top_products)

# Segment customers based on total spending
marketing_data['SpendingSegment'] = pd.qcut(marketing_data['TotalSpending'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Analyze customer segments
segment_means = marketing_data.groupby('SpendingSegment')[spending_columns].mean()
print("Average Spending by Segment:")
print(segment_means)
```

```python
#### Marketing Campaign Effectiveness
###Assessing the effectiveness of marketing campaigns using acceptance rates
#campaign acceptance columns
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']

# calculating the acceptance rates
acceptance_rates = marketing_data[campaign_columns].mean() * 100

# Plot of the acceptance rates
plt.figure(figsize=(12, 7))

sns.barplot(x=acceptance_rates.index, y=acceptance_rates.values, palette='coolwarm')
plt.xlabel('Campaign')
plt.ylabel('Acceptance Rate (%)')
plt.title('Acceptance Rates of Marketing Campaigns')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

# Print acceptance rates for review
print("Acceptance Rates for Each Campaign:")
print(acceptance_rates)
```

```python
#Checking if there is a correlation between the acceptance of different campaigns.
correlation_matrix = marketing_data[campaign_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Campaign Acceptances')
plt.show()
```

```python
#predictive analysis
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Example feature columns and target variable
feature_columns = ['Age', 'Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
target_column = 'Response'

# Features and target
X = marketing_data[feature_columns]
y = marketing_data[target_column]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train the model
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Responded', 'Responded'], yticklabels=['Not Responded', 'Responded'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
### visualization
#Using a heatmap of the correlation matrix to see if Age, Education, and Income are correlated.

# Select relevant columns
relevant_columns = ['Age', 'Education', 'Income']
df_relevant = marketing_data[relevant_columns]

# Calculate the correlation matrix
correlation_matrix = df_relevant.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Age, Education, and Income')
plt.show()
```

```python
####Analyzing average income by education level to see if higher education correlates with higher income.

#since education is a catetegorical variable
marketing_data['Education'] = marketing_data['Education'].astype('category')

# Calculate average income by education level
average_income_by_education = marketing_data.groupby('Education')['Income'].mean().reset_index()

# Sort by average income for better visualization
average_income_by_education = average_income_by_education.sort_values(by='Income', ascending=False)

# Plot average income by education level
plt.figure(figsize=(10, 6))
sns.barplot(x='Education', y='Income', data=average_income_by_education, palette='viridis')
plt.xlabel('Education Level')
plt.ylabel('Average Income')
plt.title('Average Income by Education Level')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
